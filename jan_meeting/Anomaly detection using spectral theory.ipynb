# Laplacian matrix of the graph

L = D - A

```python
import numpy
import scipy
import sys
from scipy.sparse import coo_matrix, identity, SparseEfficiencyWarning
from numpy import array
from scipy import linalg
import math
import operator

def compute_laplacian_matrix(input_graph_name, input_data_name):
    L = []
    node_ids = {}
    ID = 0
    
    input_data = open(input_data_name, 'r')

    for line in input_data:
        line = line.rstrip()
        vec = line.rsplit(',')

        vertex = vec[0]
        
        if vertex not in node_ids:
            node_ids[vertex] = ID
            ID = ID + 1

    input_data.close()

    for i in range(0, len(node_ids)):
        L.append([])
        for j in range(0, len(node_ids)):
            L[i].append(0)

    input_graph = open(input_graph_name, 'r')
    
    for line in input_graph:
        line = line.rstrip()
        vec = line.rsplit(',')

        v_one = vec[0]
        v_two = vec[1]
        
        id_one = node_ids[vec[0]]
        id_two = node_ids[vec[1]]
        
        L[id_one][id_two] = 1
        L[id_two][id_one] = 1 
    input_graph.close()
    
    for i in range(0, len(node_ids)):
        d = 0
        for j in range(0, len(node_ids)):
            d = d + L[i][j]
            L[i][j] = -1*L[i][j]
        
        L[i][i] = d

    return numpy.array(L), node_ids       
```

```python
L, node_ids = compute_laplacian_matrix(data_path+"traffic.graph", data_path+"traffic_0.data")
```

# Reading node values

```python
def read_values(input_data_name, ids):
    F = []
    input_data = open(input_data_name, 'r')

    for i in range(0, len(ids)):
        F.append(0)

    for line in input_data:
        line = line.rstrip()
        vec = line.rsplit(',')

        vertex = vec[0]
        value = float(vec[1])

        F[ids[vertex]] = value

    input_data.close()

    return numpy.array(F)  
```

# Spectral decomposition of the Laplacian matrix

```python
def compute_eigenvectors_and_eigenvalues(L):
    lamb,U = linalg.eig(L)

    return U, lamb

U, lamb_str = compute_eigenvectors_and_eigenvalues(L)
```

# Graph Fourier

```python
def graph_fourier(F, U):
    lambdas = []

    for i in range(0, len(U)):
        lambdas.append(numpy.dot(F, U[:,i]))    
    
    lambdas = numpy.array(lambdas)

    return lambdas
```

# Gaussian distribution

```python
import sys

def gaussian_dist(x, mu, std):
    prob = math.exp(float(-math.pow(x - mu, 2.)) / (2 * math.pow(std, 2.)))
    
    if prob == 0:
        prob = sys.float_info.epsilon
        
    return prob
```

# Computing p-values for eigenvectors and each snapshot

```python
p_values = []

for i in range(0, num_snapshots): 
    p_values.append([])
    
    for j in range(0, len(lambdas)):
        p_value = gaussian_dist(lambda_values[i][j], mean_lambdas[j], std_lambdas[j])
        p_values[i].append(p_value)
```

# Showing p-values for each snapshot (product)

log P(Graph | Model) = log P(lambda_1 | Model) + log P(lambda_2 | Model) ...

We assume probabilities are independent and low p-value represent outliers.

```python
x = []
y = []
for i in range(0, num_snapshots):
    sum_log_p_values = 0
    for j in range(0, len(lambdas)):
        sum_log_p_values = sum_log_p_values + math.log(p_values[i][j])
        
    x.append(i+1)
    y.append(sum_log_p_values)
        
plt.clf()
plt.scatter(x,y, color='blue',s=5,edgecolor='none')
plt.savefig("p_values.png")
Image(filename='p_values.png')
```

# Showing p-values for each snapshot (minimum)

```python
x = []
y = []
for i in range(0, num_snapshots):
    min_log_p_value = 0
    for j in range(0, len(lambdas)):
        if min_log_p_value >= math.log(p_values[i][j]):
            min_log_p_value = math.log(p_values[i][j])                               
        
    x.append(i+1)
    y.append(min_log_p_value)
        
plt.clf()
plt.scatter(x,y, color='red',s=5,edgecolor='none')
plt.savefig("p_values.png")
Image(filename='p_values.png')
```

# Adding synthetic anomaly randomly swaping node values

```python
import random

def swap_values(F, num):
    for i in range(0, num):
        p_one = random.randint(0, len(F)-1)
        p_two = random.randint(0, len(F)-1)
        value = F[p_one]
        F[p_one] = F[p_two]
        F[p_two] = value
    
def compute_sum_min_log_p_value(lambdas, mean_lambdas, std_lambdas):
    p_values_anomaly = []
    for j in range(0, len(lambdas)):
        p_value = gaussian_dist(lambdas[j], mean_lambdas[j], std_lambdas[j])
        p_values_anomaly.append(p_value)
    
    sum_log_p_values = 0
    for j in range(0, len(lambdas)):
        sum_log_p_values = sum_log_p_values + math.log(p_values_anomaly[j])
        
    min_log_p_value = 0
    for j in range(0, len(lambdas)):
        if min_log_p_value > math.log(p_values_anomaly[j]):
            min_log_p_value = math.log(p_values_anomaly[j])
            
    return sum_log_p_values, min_log_p_value

file_name = data_path + "traffic_0.data"
F = read_values(file_name, node_ids)

x = []
y_sum = []
y_min = []

for num in range(0, 500, 50):
    if num > 0:
        swap_values(F, 50)
        lambdas = graph_fourier(F, U)
        sum_p, min_p = compute_sum_min_log_p_value(lambdas, mean_lambdas, std_lambdas)
        x.append(num)
        y_sum.append(sum_p)
        y_min.append(min_p)
    
print y_sum
print y_min

plt.clf()
plt.bar(x,y_sum)
plt.xlabel('#swaps')
plt.ylabel('sum log p-value')
plt.savefig("sum_p_values_swaps.png")
Image(filename='sum_p_values_swaps.png')
```
