```python
import numpy
import scipy
from scipy.sparse import coo_matrix, identity, SparseEfficiencyWarning
from numpy import array
from scipy import linalg
import math

def Laplacian(G):
    L = []
    
    for i in range(0, len(G.nodes())):
        L.append([])
        for j in range(0, len(G.nodes())):
            L[i].append(0)
    
    for edge in G.edges():
        L[edge[0]-1][edge[1]-1] = 1
        L[edge[1]-1][edge[0]-1] = 1 
    
    for i in range(0, len(G.nodes())):
        d = 0
        for j in range(0, len(G.nodes())):
            d = d + L[i][j]
            L[i][j] = -1*L[i][j]
        
        L[i][i] = d

    return numpy.array(L)

print Laplacian(graph)
```

```python
def spectral_dec(L):
    lamb,U = linalg.eig(L)

    return U, lamb

U, lamb_str = spectral_dec(Laplacian(graph))

for i in range(0, len(U)):
    print"u =", U[:,i]
    print "lambda=", lamb_str[i]
```

```python
def graph_fourier(F, U):
    lambdas = []

    for i in range(0, len(U)):
        lambdas.append(numpy.dot(F, U[:,i]))    
    
    lambdas = numpy.array(lambdas)

    return lambdas

lamb_val = graph_fourier(F, U)

for i in range(0, len(U)):
    print"u =", U[:,i]
    print "lambda=", lamb_val[i]
    print "lambda*u", lamb_val[i] * U[:,i]
```
